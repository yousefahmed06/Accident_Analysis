# **Data Cleaning**


Fixing quality issues

```python
df_clean=df.copy()
df.info()
```

A- Define

  fill null values 'Road_Surface_Conditions' 'Road_Type' 'Weather_Conditions' and 'Urban_or_Rural_Area'

B- Code

```python
df_clean['Road_Surface_Conditions'].fillna('Dry',inplace=True)
df_clean['Road_Type'].fillna('Other',inplace=True)
df_clean['Weather_Conditions'].fillna('Other',inplace=True)
df_clean['Urban_or_Rural_Area'].fillna('Unallocated',inplace=True)
```

C- test

```python
df_clean.info()
```

A- Define

  change 'Accident Date'to datetime

B- Code

```python
df_clean['Accident Date'] = pd.to_datetime(df_clean['Accident Date'], format='%d-%m-%Y')
```
